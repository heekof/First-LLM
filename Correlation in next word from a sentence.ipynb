```python
import torch
import torch.nn as nn
import torch.optim as optim

class EnhancedContextWindowModel(nn.Module):
    def __init__(self, vocab_size, embedding_dim, hidden_dim):
        super(EnhancedContextWindowModel, self).__init__()
        self.embeddings = nn.Embedding(vocab_size, embedding_dim)
        # A linear layer that transforms the input embeddings to a hidden layer
        self.fc1 = nn.Linear(embedding_dim, hidden_dim)
        # A second linear layer that maps from the hidden layer to the vocabulary size
        self.fc2 = nn.Linear(hidden_dim, vocab_size)
        # Dropout layer
        self.dropout = nn.Dropout(0.5)

    def forward(self, context_words):
        # Embedding layer
        embedded = self.embeddings(context_words)  # Shape: [context_size, embedding_dim]
        # Average embeddings of context words
        context_embedding = torch.mean(embedded, dim=0)  # Shape: [embedding_dim]
        # First fully connected layer with ReLU activation
        hidden = torch.relu(self.fc1(context_embedding))
        # Apply dropout
        hidden = self.dropout(hidden)
        # Output layer
        out = self.fc2(hidden)
        log_probs = torch.log_softmax(out, dim=0)
        return log_probs

# Example usage
vocab_size = len(word_to_number.keys())  # Example, set this to your actual vocabulary size
embedding_dim = 5
hidden_dim = 20  # Hidden layer size

# Training setup
big_model = EnhancedContextWindowModel(vocab_size, embedding_dim, hidden_dim)
loss_function = nn.NLLLoss()
optimizer = optim.Adam(big_model.parameters(), lr=0.01)

training_data = mapped_data

# Assuming you have a training_data variable set up as before
epochs = 1000
for epoch in range(epochs):
    total_loss = 0
    for context, target in training_data:
        context_var = torch.tensor(context, dtype=torch.long)
        target_var = torch.tensor([target], dtype=torch.long)

        # Forward pass
        log_probs = big_model(context_var)

        # Compute and print loss
        loss = loss_function(log_probs.view(1, -1), target_var)
        total_loss += loss.item()

        # Backward and optimize
        optimizer.zero_grad()
        loss.backward()
        optimizer.step()

    if epoch % 10 == 0:
        print(f'Epoch {epoch}, Loss: {total_loss}')

# To get embeddings for a word
word_idx = 4  # Example word index
embedding = big_model.embeddings(torch.tensor([word_idx]))
print(embedding)
```

```python
import torch

def find_closest_word(embedding, embeddings_matrix, idx2word):
    """
    Find the closest word for a given embedding.
    
    Args:
    - embedding (torch.Tensor): The embedding vector.
    - embeddings_matrix (torch.Tensor): Matrix of embeddings from the model.
    - idx2word (dict): Dictionary mapping indices to words.
    
    Returns:
    - str: The closest word to the given embedding.
    """
    # Compute distances between the given embedding and all embeddings in the matrix
    distances = torch.norm(embeddings_matrix - embedding, dim=1)
    # Find the index of the minimum distance
    closest_idx = torch.argmin(distances).item()
    # Return the word corresponding to the closest index
    return idx2word[closest_idx]

```
